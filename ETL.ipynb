```python
def limpiar_telefono(lista_telefonos):
  lista_numeros_limpios = []

  for numero in lista_telefonos:
    numero_limpio = numero.replace("-","").replace(" ","").strip()
    lista_numeros_limpios.append(numero_limpio)

  #lista_numeros_limpios = ["987654321","999888777","987987987"]
  return lista_numeros_limpios
```

Recibimos la lista de ventas del día de un asesor comercial. La regla de la empresa es:

Si la venta es mayor a $1000, gana el 10% de comisión.

Si la venta está entre  500y 1000, gana el 5%.

Si la venta es menor a $500, no gana comisión (0%).

Queremos una función que calcule cuánto dinero extra se llevará el asesor hoy.


[ ]


```python
ventas_asesor_A = [1200, 300]
```

```python
def calcule_pago_comisiones(lista_ventas):
  total_comision = 0 #Inicializar el acumulador de comisiones

  for venta in lista_ventas:
    if venta > 1000:
      total_comision = total_comision + venta*0.1
    elif venta >= 500:
      total_comision = total_comision + venta*0.05

  return total_comision
```

```python
comision_asesor_A = calcule_pago_comisiones(ventas_asesor_A)
print(comision_asesor_A)
```

```python
def comision_venta(lista_venta):
  total_comision = 0
  for venta in lista_venta:
    if venta > 1000:
        total_comision = total_comision + venta * 0.1
    elif venta > 500 and venta < 1000:
        total_comision = total_comision + venta * 0.05

  return total_comision

lista_venta = [600, 1500.5, 501, 360]
comision = comision_venta(lista_venta)
print(f"El asesor ganará ${round(comision,2)} de comisión.")
```

```python
def extraer_montos(lista_datos):
    """
    Recorre una lista de diccionarios y extrae únicamente los valores de 'monto_usd'.
    """
    lista_montos = []
    for registro in lista_datos:
        # Extraemos el valor asociado a la llave "monto_usd" en cada ciclo
        monto = registro["monto_usd"]
        lista_montos.append(monto)
    return lista_montos
```

```python
# Prueba de la Parte 1
montos_extraidos = extraer_montos(respuesta_api)
print("1. Montos extraídos de la API:", montos_extraidos)
```

```python
def convertir_a_soles(monto):
    """
    Convierte un monto en dólares a soles (tipo de cambio 3.8).
    Incluye un mecanismo de seguridad para evitar colapsos por datos corruptos.
    """
    # IMPLEMENTACIÓN DEL TRY/EXCEPT:
    # Se utiliza este bloque porque sabemos que la API envía ocasionalmente texto
    # en lugar de números (ej. "Error_Red"). Si intentamos multiplicar un texto,
    # Python lanzará un TypeError y detendrá todo el sistema. El try/except
    # 'atrapa' ese error, reporta el problema y permite que el código siga fluyendo.
    try:
        conversion = monto * 3.8
        return round(conversion, 2)
    except Exception as e:
        print(f"\n⚠️ Alerta: Dato corrupto detectado. Detalle: [{e}]")
        return 0
```
